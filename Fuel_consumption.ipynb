# US Fuel consumption

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Fuel consumption for water heating

```python
dfw=pd.read_excel("water heating.xlsx")
dfw
```

```python
dfw.info()
```

```python
dfw.describe()
```

```python
dfw.isna().sum()
```

```python
dfw.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dfw1=dfw["Electricity"].sum()
dfw2=dfw["NaturalGas"].sum()
dfw3=dfw["Propane"].sum()
dfw4=dfw["Fueloil/Kerosene"].sum()
```

```python
label=["Electricity", "NaturalGas", "Propane", "Fueloil/Kerosene"]
colours = {"Teal", "Magenta", "Red", "Orange"}
plt.pie([dfw1,dfw2,dfw3,dfw4],labels=label, autopct='%1.2f%%', colors = colours)
plt.title("Different Fuel types used for water heating")
```

```python
plt.bar(dfw["Climate"], dfw["Total"])
plt.bar(dfw["Climate"], dfw["Total"], color='teal', width=0.9, linewidth=2)
plt.xlabel('Climate')
plt.ylabel('Fuel usage')
plt.title('Bar plot of climate wise fuel usage for water heating')
plt.xticks(rotation=60, ha='right', fontsize=10)
```

```python
dfw_corr = pd.DataFrame(dfw.corr().unstack().sort_values(ascending=False)['Total'],columns=['correlation to the target'])
dfw_corr.style.background_gradient(cmap=sns.light_palette('green',as_cmap=True))
```
